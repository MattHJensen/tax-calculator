### Import taxcalc package and pandas

```python
from taxcalc import *
import pandas as pd
```

### Create baseline calculator from new policy and records class object

```python
policy_x = Policy()
recs_x = Records()
calc_x = Calculator(policy=policy_x, records=recs_x)
```

### Create reform calculator

```python
policy_y = Policy()
# Create reform dictionary
reform = {
    2016: {
        '_CTC_new_rt': [1],
        '_CTC_new_c': [2000],
        '_CTC_new_prt': [0.05],
        '_CTC_new_ps': [[75000, 110000, 55000, 75000, 75000, 55000]]
    }
}
# Implement reform
policy_y.implement_reform(reform)
recs_y = Records()
calc_y = Calculator(policy=policy_y, records=recs_y)
```

### Advance to 2016 and run the calculators

```python
calc_x.advance_to_year(2016)
calc_x.calc_all()
calc_y.advance_to_year(2016)
calc_y.calc_all()
```

### Create and export differences table

```python
diffs_table = utils.create_difference_table(recs_x, recs_y, groupby='small_income_bins')
diffs_table.to_csv('diffs_table.csv')
diffs_table
```

### Create and export one-year diagnostic table

```python
diagnotic_table = utils.create_diagnostic_table(calc_y)
diagnotic_table.to_csv('diagnostic_table.csv')
diagnotic_table
```

### Create and export 10-year diagnostic table

```python
ten_year_table = utils.multiyear_diagnostic_table(calc_y, 10)
ten_year_table.to_csv('ten_year_table.csv')
ten_year_table
```
